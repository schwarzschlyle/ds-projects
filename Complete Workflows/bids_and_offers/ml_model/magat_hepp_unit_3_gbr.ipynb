```python
X = unit_3_df[['U3_GVposition', 'Maris_level',
       'Magat_level', 'Tailrace_level', 'net_head']]
Y = unit_3_df[['U3_ActivePower']]


validation_size = 0.995

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size= validation_size,
                                                    random_state=42)
```

```python
# Models to try
models = []
models.append(('LR', LinearRegression()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('SVR', SVR()))
models.append(('MLP', MLPRegressor()))
models.append(('ABR', AdaBoostRegressor()))
models.append(('GBR', GradientBoostingRegressor()))
models.append(('RFR', RandomForestRegressor()))
models.append(('ETR', ExtraTreesRegressor()))
models.append(('XGB', XGBRegressor()))
```

```python
# Model evaluation function

def evaluate_models(models,metric,to_plot=False):
    num_folds = 10
    seed = 42
    # scikit is moving away from mean_squared_error. 
    # In order to avoid confusion, and to allow comparison with other models, 
    # we invert the final scores
    scoring = {'mse':'neg_mean_squared_error','mae':'neg_mean_absolute_error','r2':'r2'}
    metric_function = {'mse':mean_squared_error,'mae':mean_absolute_error,'r2':r2_score}
    title = {'mse':'Mean Squared Error','mae':'Mean Absolute Error','r2':'R-Squared'}

    # Ready the output lists
    names = []
    kfold_results = []
    test_results = []
    train_results = []

    for name, model in models:
        names.append(name)

        ## K Fold analysis:
        kfold = KFold(n_splits=num_folds)
        #converted mean square error to positive. The lower the beter
        cv_results = -1* cross_val_score(model, X_train, Y_train, 
                                         cv=kfold, scoring=scoring[metric])
        kfold_results.append(cv_results)

        # Full Training period
        res = model.fit(X_train, Y_train)
        train_result = metric_function[metric](res.predict(X_train), Y_train)
        train_results.append(train_result)

        # Test results
        test_result = metric_function[metric](res.predict(X_test), Y_test)
        test_results.append(test_result)

    if to_plot:
        fig = plt.figure()
        ind = np.arange(len(names)-2)  # the x locations for the groups
        width = 0.35                   # the width of the bars

        plt.rcParams['axes.facecolor']='white'

        # Plot into bar charts
        fig.suptitle(title[metric])
        ax = fig.add_subplot(111)
        plt.bar(ind - width/2, train_results[0:1] + train_results[3:],  width=width, label='Train Error')
        plt.bar(ind + width/2, test_results[0:1] + test_results[3:], width=width, label='Test Error')

        fig.set_size_inches(10,4)
        fig.patch.set_facecolor('white')
        plt.grid(color='lightgrey', linestyle='-.', linewidth=0.7)
        plt.legend()

        ax.spines['right'].set_color('lightgrey') 
        ax.spines['left'].set_color('lightgrey') 
        ax.spines['top'].set_color('lightgrey') 
        ax.spines['bottom'].set_color('lightgrey') 
        ax.set_xticks(ind)
        ax.set_xticklabels(names[0:1] + names[3:])
        plt.show()
    
    output = pd.DataFrame({'Model':names,'Train Result': train_results,
                         'Test Result': test_results})
    output = output.set_index('Model')
    return output
```

```python
mse_results = evaluate_models(models,'mse',to_plot= False)
mae_results = evaluate_models(models,'mae',to_plot= True)
r2_results = evaluate_models(models,'r2',to_plot= False)
```
